## Data Extraction 

#### Importing the csv file 

```python
# import basic libraries 
import pandas as pd 
import numpy as np
```

## Data Cleaning 

#### Input a title 

```python
# import csv file 
# note that default behaviour is to infer the column names, identical to header = 0, default behaviour assumes 
# that the first line(line 0) is the header
# in this case there are no headers, but we can use the attribute names to make the headers 
expense_data = pd.read_csv('expenses.csv', names=['Expense Name', 'Category', 'Amount'])
expense_data.head()
```

#### Getting to know more about the data

```python
# finding the data types of each attribute
expense_data.dtypes
```

```python
# similar to above, but also gives us no. of non-null entries 
expense_data.info()
```

```python
# finding the shape of the df (row x column)
expense_data.shape
```

```python
# gives us some basic stats - but only for numeric variables 
expense_data.describe()
```

From this we can infer that:
* The average amount spent on an expense is about $9

* The most expensive expense costs $20 

* The cheapest expense costs $1

## Data Analysis 

#### Importing the necessary libraries 

```python
import seaborn as sb 
import matplotlib.pyplot as plt 
sb.set()
```

#### Making a histogram

```python
f = plt.figure(figsize=(24,12))
sb.histplot(data=expense_data)
```

#### Making a violin plot 

```python
f = plt.figure(figsize=(24,12))
sb.violinplot(data=expense_data, orient="h")
```
